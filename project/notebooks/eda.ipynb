```python
target_counts =df["y"].value_counts()
print(target_counts)
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
target_counts.plot(kind="bar", ax=ax, edgecolor="white")
ax.set_title("Распределение целевой переменной y", fontsize=13)
ax.set_xlabel("Оформил вклад")
ax.set_ylabel("Количество клиентов")
ax.set_xticklabels(["Нет", "Да"], rotation=0)
plt.tight_layout()
plt.savefig("../artifacts/figures/fig_target_distribution.png", dpi=150)
plt.show()
```
